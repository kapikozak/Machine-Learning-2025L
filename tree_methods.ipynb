#Equivalence Analysis

##RSS Gain Maximization

The parent $P^*$, the feature $X_j^*$ and the threshold $t^*$ are selected as
\begin{equation}
  \mathrm{argmax}_{P, X_j, t}\big(\sum_{i: x_i \in P}(y_i - \overline y_P)^2 - \sum_{m = 1}^2 \big[\sum_{i: x_i \in R_m(X_j, t)} (y_i - \overline y_{R_m})^2 \big]\big)
\end{equation}
in terms of $P$, $X_j$ and $t$.

For all $P$, $X_j$ and $t$
\begin{align}
  \sum_{i: x_i \in P}(y_i - \overline y_P)^2
  &= \sum_{i: x_i \in R_1(X_j, t)} (y_i - \overline y_P + \overline y_{R_1} - \overline y_{R_1})^2\\
  &\quad + \sum_{i: x_i \in R_2(X_j, t)} (y_i - \overline y_P + \overline y_{R_2} - \overline y_{R_2})^2\\
  &= \sum_{m = 1}^2 \big[\sum_{i: x_i \in R_m(X_j, t)} (y_i - \overline y_{R_m})^2\big]\\
  &\quad + 2\sum_{m = 1}^2 \big[(\overline y_{R_m} - \overline y_P)\sum_{i: x_i \in R_m(X_j, t)} (y_i - \overline y_{R_m})\big]\\
  &\quad + \sum_{m = 1}^2 |R_m| (\overline y_{R_m} - \overline y_P)^2.
\end{align}
The second term vanishes, becouse for $m = 1, 2$
\begin{equation}
  \sum_{i: x_i \in R_m(X_j, t)} (y_i - \overline y_{R_m}) = 0.
\end{equation}
Denoting $n = |P|$, $n_1 = |R_1|$, $n_2 = n - n_1$ and using that
\begin{equation}
  \overline y_P = \frac{n_1 \overline y_{R_1} + n_2 \overline y_{R_2}}{n},
\end{equation}
the last term can be written as
\begin{align}
  \sum_{m = 1}^2 |R_m| (\overline y_{R_m} - \overline y_P)^2
  &= \frac{n_1n_2^2}{n^2}(\overline y_{R_1} - \overline y_{R_2})^2 + \frac{n_1^2n_2}{n^2}(\overline y_{R_1} - \overline y_{R_2})^2\\
  &= \frac{n_1n_2}{n}(\overline y_{R_1} - \overline y_{R_2})^2.
\end{align}
So that
\begin{equation}
  \sum_{i: x_i \in P}(y_i - \overline y_P)^2 - \sum_{m = 1}^2 \big[\sum_{i: x_i \in R_m(X_j, t)} (y_i - \overline y_{R_m})^2\big] = \frac{n_1n_2}{n}(\overline y_{R_1} - \overline y_{R_2})^2.
\end{equation}

What we essentially do in each split is looking for
\begin{equation}
  \mathrm{argmax}_{P, X_j, t} \frac{n_1n_2}{n}(\overline y_{R_1} - \overline y_{R_2})^2.
\end{equation}

##Local RSS Minimization

We select $P^*$, $X_j^*$ and $t^*$ that minimizes sum of squared errors in the two resulting child nodes i.e.
\begin{equation}
  \mathrm{argmin}_{P, X_j, t}\sum_{m = 1}^2 \big[\sum_{i: x_i \in R_m(X_j, t)} (y_i - \overline y_{R_m})^2\big].
\end{equation}

Let us assume that we have two parent nodes to choose from, consisting of one-feature elements with values $a_1 < a_2$ and $b_1 < b_2 < b_3$, respectively. Let us further assume that
\begin{equation}
  \frac{2}{3}\big(b_1 - \frac{b_2 + b_3}{2}\big)^2 - \frac{1}{2}(a_1 - a_2)^2 = C > 0.
\end{equation}

We want to compare split $s_a$ in the first subtree at threshold
\begin{equation}
  a_1 + \frac{a_2 - a_1}{2}
\end{equation}
with split $s_b$ in the second subtree at threshold
\begin{equation}
  b_1 + \frac{b_2 - b_1}{2}.
\end{equation}

The RSS Gain Maximization criterion would prefer $s_b$, since it provides a larger gain, i.e. $C > 0$.

On the other hand, for the two-element subtree, the local RSS formula yields $0$, while for the three-element subtree we obtain
\begin{align}
  f(b_1, b_2, b_3)
  &= (b_1 - b_1)^2 + \big(b_2 - \frac{b_2 + b_3}{2}\big)^2\\
  &\quad + \big(b_3 - \frac{b_2 + b_3}{2}\big)^2 \\
  &= 2\big(\frac{b_2 - b_3}{2}\big)^2 \\
  &> 0.
\end{align}

Hence, the Local RSS Minimization criterion would prefer split $s_a$.

When $P$ is fixed, however, we have
\begin{align}
  &\mathrm{argmin}_{X_j, t} \sum_{m = 1}^2 \big[ \sum_{i: x_i \in R_m(X_j, t)}(y_i - \overline y_{R_m})^2 \big] \\
  &\quad = \mathrm{argmax}_{X_j, t}
  \big(\sum_{i: x_i \in P}(y_i - \overline y_P)^2 - \sum_{m = 1}^2 \big[\sum_{i: x_i \in R_m(X_j, t)}(y_i - \overline y_{R_m})^2\big]\big).
\end{align}

Therefore, for fixed $P$, the two criteria are equivalent. However, they are not equivalent when optimizing jointly over $P$, $X_j$, and $t$.

In the standard unpruned greedy tree-growing procedure, both formulations lead to the same choice of split within a selected parent node. The difference matters only when comparing candidate parent nodes directly, for example when limiting the number of leaves.

##Total RSS Minimization

Let $L(T(P, X_j, t))$ denote the set of all leaves in tree T that was created after splitting leaf $P$ by feature $X_j$ at threshold $t$. We select $P^*$, $X_j^*$, $t^*$ as
\begin{equation}
  \mathrm{argmin}_{P, X_j, t} \sum_{m \in L(T(P, X_j, t))}\big[\sum_{i: x_i \in R_m(X_j, t)} (y_i - \overline y_{R_m})^2 \big].
\end{equation}

By subtracting sum as above from the RSS sum corresponded to the previous tree, it is a constant value, we obtain that it is equivalent to take
\begin{equation}
  \mathrm{argmax}_{P, X_j, t}\big(\sum_{i: x_i \in P}(y_i - \overline y_P)^2 - \sum_{m = 1}^2 \big[\sum_{i: x_i \in R_m(X_j, t)} (y_i - \overline y_{R_m})^2 \big]\big).
\end{equation}
This is the same optimalization problem as the first one.

Design and conduct a Python experiment to determine which one of these formulations is implemented in `scikit-learn` in `DecisionTreeRegressor`. Present numerical results and plots to support your conclusion.



All we have to do is to evaluate the example from above.

We will provide data that would lead to the $2/3$ split with appropriate assumptions in the first divison i.e. $x^T = [0, 1, 2, 3, 4]$, $y^T = [100, 110, 120, 0.1, 0.2]$. Then
\begin{equation}
  \frac{2}{3}\big(100 - 115\big)^2 - \frac{1}{2}(0.1)^2 > 0.
\end{equation}
The first split is not the problem for us, as in this case $P$, the root, is fixed.

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor, export_text

x = np.array(range(5)).reshape(-1, 1)
y = np.array([100, 110, 120, 0.1, 0.2])

dt_pruned = DecisionTreeRegressor(random_state=42, max_leaf_nodes=3)
dt_pruned.fit(x, y)
tree_rules = export_text(dt_pruned, feature_names=('x',))
print(tree_rules)
```
